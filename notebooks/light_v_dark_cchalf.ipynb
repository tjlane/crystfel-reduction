```python
import pandas as pd
from matplotlib import pyplot as plt
from pathlib import Path
```

```python
base_path = Path("/Users/tjlane/Desktop/final_stats")
dataset = "mmCPD_zeroDTT_10us_21ph-xsection"
states = ["dark", "light"]
```

```python
fig, axes = plt.subplots(2, 1, sharex=True, figsize=(4,6))


for state in states:
    fn = base_path / Path(f"{dataset}_{state}_stats_by_shell.csv")
    df = pd.read_csv(fn)

    ax = axes[0]
    ax.grid(True)
    ax.plot(df["Center 1/nm"], df["e       CC"], "-o")
    ax.set_ylabel(r"$CC_\mathrm{1/2}$")

    ax = axes[1]
    ax.grid(True)
    ax.plot(df["Center 1/nm"], df["Rsplit/%"], "-o")
    ax.set_xticklabels(df["d(A)"])
    ax.set_ylabel(r"$R_\mathrm{split}$")

ax.set_xlabel("Resolution (Ã…)")

axes[0].legend(states)
plt.subplots_adjust(wspace=None, hspace=None)
plt.tight_layout()
plt.savefig(f"{dataset}_cc-stats.pdf")
plt.show()
```
